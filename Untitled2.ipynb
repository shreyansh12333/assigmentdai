```python
#UNIVARIATE ANALYSIS
num_col = ['Age','Height (cm)', 'Weight (kg)','Workout Duration (mins)', 'Calories Burned','Heart Rate (bpm)', 'Steps Taken', 'Distance (km)','Sleep Hours', 'Daily Calories Intake','Resting Heart Rate (bpm)']
len(num_col)
```

```python
import warnings 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
warnings.filterwarnings('ignore') 
plt.figure(figsize = (15,12))
for i, cols in enumerate(num_col):
    plt.subplot(4,3,i+1)
    sns.histplot(data = data,x = cols,kde = True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12,10))
#performing outlier detection
for i, cols in enumerate(num_col):
    plt.subplot(4,3,i+1)
    sns.boxplot(data = data,x = cols)
plt.tight_layout()
plt.show()
```

```python
#skewness
data[num_col].skew()
```

```python
#UNIVARIATE ANALYSIS ON CATEGORICAL COLUMNS
cat_col = ['Gender','Workout Type','Workout Intensity','Mood Before Workout','Mood After Workout']
plt.figure(figsize = (12,6))

for i,cols in enumerate(cat_col):
    plt.subplot(2,3,1+i)
    sns.countplot(data=data,x=cols)
plt.tight_layout()
plt.show()
```

```python
#AGE VS WORKOUT DURATION
sns.histplot(data=data,x= 'Age',y='Workout Duration (mins)',cmap='viridis')
sns.color_palette("viridis", as_cmap=True)
"""
Tend to have higher workout durations (~45-60 mins on average).
Likely due to greater energy levels, fitness enthusiasm, or structured workout routines.
Shows a wider range of workout durations, but the median is slightly lower (~30-45 mins).
Likely due to time constraints (work, family) or a shift towards moderate-intensity workouts.
Generally engage in shorter workouts (~20-40 mins).
Likely due to a preference for lower-intensity exercises (e.g., walking, light strength training).


"""
```

```python
sns.boxplot(data=data,x = 'Workout Intensity',y='Sleep Hours')
"""
Individuals with moderate workout intensity tend to have consistent sleep durations, around 6-8 hours on average.
This suggests that moderate exercise helps maintain a stable sleep pattern.
High-intensity workout participants show greater variability in sleep hours.
Some individuals sleep longer (~9 hours), possibly for recovery, while others sleep less (~5 hours), possibly due to overtraining effects.
Those engaging in low-intensity workouts tend to have lower median sleep hours (~6 hours).
This might indicate that lower activity levels are linked to less restful or shorter sleep durations
"""
```

```python
data['heart_rate_difference'] = data['Heart Rate (bpm)'] - data['Resting Heart Rate (bpm)']
```

```python
#HEART RATE DIFFERENCES DURING WORKOUTS
plt.figure(figsize = (8,6))
sns.barplot(data=data,x = 'Workout Type', y ='heart_rate_difference',hue='Gender')
"""
HIIT and Cardio workouts show the highest heart rate increase, indicating their intensity.
Strength training and cycling have moderate heart rate increases, likely due to resistance-based efforts and pacing
Males and females show similar trends across workout types, but males tend to have slightly higher heart rate increases in HIIT and Strength workouts.
Females show more variation in heart rate responses, possibly due to different fitness levels or training styles.
Cycling and Strength training show lower heart rate increases compared to HIIT and Cardio, suggesting lower cardiovascular demand in these exercises.
This aligns with expectations, as endurance-based workouts (Cycling) often allow the heart rate to stabilize.
"""
```

```python
sns.kdeplot(data=data,x = 'Workout Duration (mins)', y = 'Heart Rate (bpm)',fill=True)
"""
There are high-density regions where heart rates cluster around 120-160 bpm for workout durations between 20 to 60 minutes.
This suggests that most users maintain a moderate-to-high heart rate within this duration range.
As workout durations exceed 60 minutes, heart rates tend to stabilize around 110-140 bpm.
This could indicate that users pace themselves during longer workouts to avoid excessive fatigue.
For workouts lasting less than 30 minutes, we observe a higher heart rate concentration above 140 bpm.
This suggests that shorter workouts often involve high-intensity exercises like HIIT, leading to elevated heart rates
"""
```
